```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# plt.show() for non-notebook users
```

In the first cell, we are simply creating data x and y to use in the provided tasks.


```python
# Create data
x = np.arange(0,10)
y = x*2
z=x**2
```

1. Instructions:
- Use plt.figure() and create a figure object, say fig.
- Add an axis to the figure canvas (fig), [left, bottom, width, height] is [0,0,1,1]. 

Hint: add_axes()
- Plot the data x,y on the axes
- Set the labels x and y
- Set the title
- Create legends as shown in the plot

```python
fig = plt.figure() # create figure obj
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,'g',label='Y = x*2')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Plot title')
axes.legend(loc=4)
```

2. Instructions 
- Create a figure object
- Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]
- Plot data(x,z) on main axes 
- Plot data(x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5 
- Replicate the figure to axes, titles, legends etc

```python
fig2 = plt.figure()
axes2 = fig2.add_axes([0,0,1,1])
axes3 = fig2.add_axes([0.1,0.6,0.2,0.2])

axes2.plot(x,z,'r',label='y_main = x_main**2')
axes2.set_xlabel('x_main')
axes2.set_ylabel('y_main')
axes2.set_title('title_main')
axes2.legend(loc=4)

axes3.plot(x,z,'b')
axes3.set_xlabel('x_inset')
axes3.set_ylabel('y_inset')
axes3.set_title('title_inset')
axes3.set_xlim(1,2)
axes3.set_ylim(0,5)
```

3. Replicate the output using subplots() and for loop  Note, the data is x and y.

```python
fig,axes=plt.subplots(nrows=1,ncols=3)
for ax in axes:
    ax.plot(x,y)
fig.tight_layout()
```
