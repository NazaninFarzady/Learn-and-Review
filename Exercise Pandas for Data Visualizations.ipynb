```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# set seed
np.random.seed(1)

# create data frame
df = pd.DataFrame(np.random.rand(1000,4), columns = ['a','b','c','d'])
```

```python
# explore dataframe, the number of enteries
df.info()

# check the first rows of dataframe
df.head()
```

```python
# plot column 'c' vs 'd' as a scatter plot
# match the color with size of data points
df.plot.scatter(x='c',y='d',c='green', s=40,figsize=(10,4))
```

```python
# create histogram of column 'd'
df['d'].plot.hist()
```

```python
# Change style sheets to set for bmh and add more bins (e.g. 50) to the above plot
plt.style.use('bmh')
df['d'].plot.hist(alpha=0.5,bins=50)
```

```python
# Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?
df.plot.scatter(x='a',y='b',s=df['d']*100, color = 'pink')
```

```python
# hex plot
df.plot.hexbin(x='a',y='b',gridsize=20,cmap='coolwarm')
```

```python
# Create a kde plot of the d column
df['d'].plot.kde()
```

```python
# Can you increase the linewidth and make the linestyle dashed. (Dont use .kde for the plot here)
df['d'].plot.density(lw=5,ls='--')
```

```python
# Create a boxplot comparing all columns in the data
# use color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray') and sym = 'r+'
color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray') 
df.plot.box(color=color, sym = 'r+')
```

```python
# Create an area plot of all the columns for just the rows up to 30. (hint: use .loc (.ix for older versions))
df.loc[10:20].plot.area(alpha=0.4)
```

```python
# Try moving the legend in the plot above and display it outside the plot, shown below
f = plt.figure()
df.loc[10:20].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
