## Matplotlib “Object Oriented” approach

object-oriented approach for more control and customization of our plots. 

https://matplotlib.org/stable/gallery/index.html

we create figure objects and then call methods or attributes from that object

Revise the previouse steps:
- Created an object (empty canvas) “fig1” – plt.figure()
- Added “axes” on “fig1” – fig1.add_axes()
- Plotted data on “axes” – axes.plot()
- Set labels and title – axes.set_xlabel/title()

we have a full control of where the plot axes are placed, and we can easily add more than one axis to the figure. Let’s learn how to create an inset plot using Object-Oriented approach!

```python
# To start with, let’s create a figure instance "fig1" and add axes to that figure:

# step 1
# create figure object (empty convas)

fig1 = plt.figure() # figure objected is created

# step 2
# add axes to the figure obj 'fig1' -- 'obj.add_axes()'
# [left,bottom,width,height] (range 0,1)

axes = fig1.add_axes([0.1,0.1,0.9,0.9])

# Step 3
# plotting the data on the set of axes we have created --axes.plot()
axes.plot(x,y,'b')

# we can add labels and titles as well
# calling methods to set labels, titles, etc --axes.set..()

axes.set_xlabel('X label') # use set_ to begin the method
axes.set_ylabel('Y label')
axes.set_title('Figure Title') #;
# put ";" if you don't want to see the line of text below!
```

- Created an object (empty canvas) "fig2".
- Added main and inset “axes” on "fig2".
- Plotted data on “main axes” and set the labels. 
- Plotted data on “inset axes”
- set the labels.

```python
# creating an empty canvas
fig2 = plt.figure()

# Creating main axes - values b/w 0 -1
axes_main = fig2.add_axes([0.1,0.1,0.9,0.9]) # [left,bottom,width,height]

# creating inset axes - values b/w 0 & 
axes_inset = fig2.add_axes([0.3, 0.5, 0.3, 0.4]) # [left,bottom,width,height]

# see the output
# change the value in axes_inset to see the difference

# main figure
axes_main.plot(x,y,'b') # 'b' means blue
axes_main.set_xlabel('X label main')
axes_main.set_ylabel('Y label main')
axes_main.set_title('Title main')


# Inset figure
axes_inset.plot(y,x,'r') # 'r' means red
axes_inset.set_xlabel('X inset label')
axes_inset.set_ylabel('Y inset label')
axes_inset.set_title('Title inset');
```

## Creating a figure and a set of subplots – “Object Oriented” approach


subplot is different then subplots (with additional s). 

subplots(): Create a figure and a set of subplots. 
A very convenient way to create layouts of subplots, including the enclosing figure object, in a single call. + <Shift+Tab> for doc string

```python
# Instead of figure(), call subplots()
# tuple unpacking to grab fig and axes

fig, axes = plt.subplots()

# adding nrows, and ncols, automatically calls the axes based on ncols and nrows
fig,axes = plt.subplots(nrows=2,ncols=2)
plt.tight_layout() # for auto adjusting


# share x and y axes ( share the units)
fig,axes = plt.subplots(nrows=2, ncols=2, sharex = True, sharey = True)
plt.tight_layout()
```

```python
# let's plot data
# Tuple unpacking to grab fig and axes
fig,axes = plt.subplots()

# Adding data to the plot using axes object.
axes.plot(x,y,'y')

# adding labels and title using axes objects
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title label')
```

```python
# Let’s create two empty canvases (fig1 and fig2) with two axes on each canvas to do some stuff

fig1, axes1 = plt.subplots(2,1) # (nrows,ncols)
fig2,axes2 = plt.subplots(1,2)

# axes1 and axes2 are arrays of axes to plot on, let's confirm their type and lenght (we can grab any one)!

type(axes1),len(axes1)
type(axes2), len(axes2)
# axes1 and axes2 are numpy arrays, so we can iterate through them using loop!
```

```python
# We can iterate through this (axes) array using for loop
for ax in axes1:
    ax.plot(x,y,'b')
    ax.set_xlabel('X label')
    ax.set_ylabel('Y label')
    ax.set_title('Title label')
fig1.tight_layout() # avoid overlapping

fig1

# Display the figure object accociated with axes1, we need it here becasue the object is created in a different cell!
```

```python
# We are also access the plots on axes1 and axes2 individually using their index values!
# access them one by one
# index 0 of axes 2
axes2[0].plot(x,y,'b')
axes2[0].set_xlabel('X blue')
axes2[0].set_ylabel('Y blue')
axes2[0].set_title('Title blue')

# index 1 of axes 2
axes2[1].plot(x,y,'r')
axes2[1].set_xlabel('X red')
axes2[1].set_ylabel('Y red')
axes2[1].set_title('Title red')

fig2.tight_layout()
fig2
```

## Decorating the figures

we can add x_label, y_label and plot_title using: axes.set_xlabel(“x_label”),
axes.set_xlabel(“y_label”) & ax.set_title(“plot_title”)

We can use the label="label_text" keyword argument when plots or other objects are added to the figure. legends() method

```python
# creating figure obj
fig3 = plt.figure()

# adding axes on fig
axes3 = fig3.add_axes([0,0,1,1])


# adding plot on axes
axes3.plot(x,x**2,label = 'X**2')
axes3.plot(x,x**3, label = 'X**3') # another plot on axes
axes3.legend() # adding legend, lower left corner
```

## Colors, linewidths, linetypes, marker styles etc

MATLAB-like syntax, 'r' means red, 'g' means green, etc can be used.

MATLAB API for selecting line styles are also supported, for example, 'r.-' means a red line with dots

The appropriate way is to use Colors with the color = parameter, by their names or RGB hex

alpha that can be used along with color to control the opacity (useful when data points are on top of each other!)

```python
fig,axes = plt.subplots() # create fig nd axes using subplot
axes.plot(x,x+1,'r.-') # red line with dot
axes.plot(x,x+2,'g--') # green dashed line
```

```python
fig, axes = plt.subplots()
axes.plot(x,x+1,color='blue', alpha=0.5) # alpha = 0.5 , half-transparent
axes.plot(x,x+2, color='#8B008B') # RGB hex color code
axes.plot(x,x+3, color='#FF8C00')
```

Let’s move on and explore little more to make the figure attractive, its important in storytelling!
- How to change line width with linewidth or lw keyword argument
- How to change the line style with linestyle or ls keyword arguments
- How to set the marker with marker and markersize keyword arguments

```python
# create fig and axes using subplots
fig, axes = plt.subplots(figsize=(10,8))

# adding plots of different colors, line widths on axes
axes.plot(x,x+1,color='red', linewidth=0.25)
axes.plot(x,x+2,color='red', linewidth=0.50)
axes.plot(x,x+3,color='red', linewidth=1.00)
axes.plot(x,x+4,color='red', linewidth=2.00)

# possible inestype options '-','-.',':','steps'
axes.plot(x,x+5,color='green', lw=5, linestyle='-')
axes.plot(x,x+6,color='green', lw=5, linestyle='-.')
axes.plot(x,x+7,color='green', lw=5, linestyle=':')

# custom dash
line,=ax.plot(x,x+8,color='black',lw=1.50)
line.set_dashes([5,10,15,10]) # format: line length,sapce, length,..

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
axes.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
axes.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
axes.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
axes.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
axes.plot(x, x+13, color="purple", lw=3, ls='-', marker='o', markersize=2)
axes.plot(x, x+14, color="purple", lw=3, ls='-', marker='o', markersize=4)
axes.plot(x, x+15, color="purple", lw=3, ls='-', marker='o', markersize=8,markerfacecolor="red")
axes.plot(x, x+16, color="purple", lw=3, ls='-', marker='s', markersize=8,markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

Matplotlib conveniently allows the control on axis

- Set the x and y limits using set_xlim and set_ylim methods
- axis('tight') for automatically getting "tightly fitted" axes ranges

```python
 # subplots again with nrows, ncols and figsize
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,3))

# Default axes range on left plot
axes[0].plot(x,y,y,x)
axes[0].set_title("default axes ranges")

# Tight axes on the middle plot
axes[1].plot(x, y, y, x)
axes[1].axis('tight')
axes[1].set_title("tight axes")

# Custom axes range on the right plot
axes[2].plot(x, y, y, x)
axes[2].set_ylim([0, 50])
axes[2].set_xlim([1, 4])
axes[2].set_title("custom axes range");
```
