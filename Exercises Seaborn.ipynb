To practice visualization with seaborn, here datasets such as tips, iris, and titanic are used.


```python
# Import libraries
# Essential imports for data analysis for visualization
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.simplefilter('ignore')
```

### loading datasets

- tips
- titanic
- iris

and set style to 'whitegrid'



```python
# Setting style and inline plotting
sns.set_style('whitegrid')

tips_ds = sns.load_dataset('tips')
titanic_ds = sns.load_dataset('titanic')
iris_ds = sns.load_dataset('iris')

# Show the head of datasets
tips_ds.head()
titanic_ds.head()
iris_ds.head()
```

```python
# make a box plot
sns.boxplot(x = 'day', y = 'tip', data = tips_ds, hue = 'sex')
```

```python
# make joint plot
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_ds, kind = 'hex')
```

```python
# make histogram plot to show the distribution
sns.histplot(titanic_ds['fare'], bins = 20, kde = False, color = 'blue')
```

```python
# box plot for titanic dataset
sns.boxplot(x = 'class',y = 'age', data = titanic_ds, hue = 'alive', palette = 'coolwarm')
```

```python
# jointplot for titanic
sns.jointplot(x = 'fare', y = 'age', data = titanic_ds)
```

```python
sns.boxplot(x = 'class',y = 'age', data = titanic_ds,  palette = 'rainbow')
```

```python
# the swarm plot using titanic dataset, use catplot. (factorplot is the older version of catplot)
#sns.factorplot(x='class', y = 'age', data = titanic_ds, col = 'sex', kind = 'swarm', palette ='rainbow')

sns.catplot(x='class',y='age',data=titanic_ds, col="sex",kind="swarm",palette='rainbow')
```

```python
# how many male and female were on titanic?
sns.countplot(x = 'sex', data = titanic_ds, hue = 'class')
```

```python
# use gridplot for iris
iris_grid = sns.PairGrid(iris_ds, hue = 'species')
iris_grid.map_diag(plt.hist)
iris_grid.map_offdiag(plt.scatter)
iris_grid.add_legend();
```

```python
# use facetgrid for titanic
titanic_grid = sns.FacetGrid(data=titanic_ds,col='sex', hue='alive')
titanic_grid.map(plt.hist,'age',alpha=0.5)
titanic_grid.add_legend()
```

```python
# use matrix plot for titanic
sns.heatmap(titanic_ds.corr(),cmap='coolwarm',annot=True)
plt.title('titanic_ds.corr()')
```
